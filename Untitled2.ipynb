```python

# Split into train and validation sets
dataset = dataset.train_test_split(test_size=0.2)
train_dataset = dataset["train"]
val_dataset = dataset["test"]
```

```python

# Preprocess the Data
from transformers import WhisperProcessor
import soundfile as sf

# Load Whisper processor
processor = WhisperProcessor.from_pretrained("openai/whisper-small")

# Import necessary libraries
import zipfile
import os
import pandas as pd
from datasets import Dataset
import soundfile as sf
from transformers import WhisperProcessor, WhisperForConditionalGeneration, Seq2SeqTrainingArguments, Seq2SeqTrainer
from google.colab import drive, files

```

```python
# Preprocess function
def preprocess(batch):
    try:
        audio_path = "/content/drive/MyDrive/tunisian_audio/" + batch["audio_file"]
        audio, sample_rate = sf.read(audio_path)
    except FileNotFoundError:
        print(f"File not found: {audio_path}")
        return None  # Skip problematic rows
    batch["input_features"] = processor(audio, sampling_rate=16000).input_features[0]
    batch["labels"] = processor.tokenizer(batch["text"]).input_ids
    return batch

```

```python
# Apply preprocessing
train_dataset = train_dataset.map(preprocess)
val_dataset = val_dataset.map(preprocess)
```

```python

# Set Up the Model and Training Arguments
from transformers import WhisperForConditionalGeneration, Seq2SeqTrainingArguments, Seq2SeqTrainer

# Load Whisper model
model = WhisperForConditionalGeneration.from_pretrained("openai/whisper-small")
```

```python
# Define training arguments
training_args = Seq2SeqTrainingArguments(
    output_dir="/content/drive/MyDrive/whisper-tunisian",
    per_device_train_batch_size=8,
    learning_rate=1e-5,
    num_train_epochs=3,
    evaluation_strategy="epoch",
    save_strategy="epoch",
    logging_steps=10,
)
```

```python
# Fine-Tune the Model
# Initialize trainer
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=processor.tokenizer,
)
```

```python
# Start training
trainer.train()
```

```python

# Save and Download the Model
# Save the model and processor
model.save_pretrained("/content/drive/MyDrive/whisper-tunisian")
processor.save_pretrained("/content/drive/MyDrive/whisper-tunisian")
```

```python
# Prepare data
data = []
for idx, row in annotations.iterrows():
    for audio_dir in audio_dirs:
        audio_path = os.path.join(audio_dir, row["file_name"])
        if os.path.exists(audio_path):
            audio, sr = sf.read(audio_path)
            duration = len(audio) / sr
            data.append({
                "ID": idx + 1,
                "wav": audio_path,
                "wrd": row["transcription"],
                "duration": duration
            })
            break  # Use the first matching audio file
```

```python
# Create DataFrame and save as CSV
df = pd.DataFrame(data)
train_csv_path = "/content/tunisian_data/train.csv"
df.to_csv(train_csv_path, index=False)
print(f"Created training CSV at: {train_csv_path}")
```

```python

# Load dataset
dataset = Dataset.from_pandas(df)
dataset = dataset.train_test_split(test_size=0.2)
train_dataset = dataset["train"]
val_dataset = dataset["test"]
```

```python

# Preprocess the Data
processor = WhisperProcessor.from_pretrained("openai/whisper-small")
processor.tokenizer.set_prefix_tokens(language="ar")

import torchaudio

def preprocess(batch):
    audio_path = batch["wav"]
    waveform, sample_rate = torchaudio.load(audio_path)

    # Convert to mono if needed
    if waveform.shape[0] > 1:
        waveform = waveform.mean(dim=0, keepdim=True)

    # Resample if needed
    if sample_rate != 16000:
        resampler = torchaudio.transforms.Resample(orig_freq=sample_rate, new_freq=16000)
        waveform = resampler(waveform)

    # Convert to numpy
    speech = waveform.squeeze().numpy()

    batch["input_features"] = processor(speech, sampling_rate=16000).input_features[0]
    batch["labels"] = processor.tokenizer(batch["wrd"]).input_ids
    return batch




"""def preprocess(batch):
    audio, sample_rate = sf.read(batch["wav"])
    batch["input_features"] = processor(audio, sampling_rate=16000).input_features[0]
    batch["labels"] = processor.tokenizer(batch["wrd"]).input_ids
    return batch"""
```

```python
import evaluate

wer_metric = evaluate.load("wer")

def compute_metrics(pred):
    pred_ids = pred.predictions
    label_ids = pred.label_ids

    # Decode predictions and labels
    pred_str = processor.batch_decode(pred_ids, skip_special_tokens=True)
    label_str = processor.tokenizer.batch_decode(label_ids, skip_special_tokens=True)

    # Compute WER
    wer = wer_metric.compute(predictions=pred_str, references=label_str)
    return {"wer": wer}

```

```python

# Fine-Tune the Model
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=processor.tokenizer,
    compute_metrics=compute_metrics,

)
```
